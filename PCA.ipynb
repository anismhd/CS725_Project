```python
import pandas as pd
import pickle as pk
import matplotlib.pyplot as plt
import tqdm
from sklearn.model_selection import train_test_split
from sklearn.datasets import fetch_california_housing
import seaborn as sns
```

```python
ground_motion = pd.read_pickle("data/processed_ground_motion_PGA.pkl")
```

```python
ground_motion_accl = ground_motion
X = ground_motion_accl.corr(method='pearson')
```

```python
f, ax = plt.subplots(figsize=(30, 30))
sns.heatmap(X, annot=False, fmt="d", linewidths=.5, ax=ax)
plt.savefig('corr_inp.png', bbox_inches='tight')
```

```python
Corr_List = list(ground_motion.columns.values)
f, ax = plt.subplots(16,16,figsize=(50, 60))
for i in range(16):
    for j in range(16):
        ax[i,j].scatter(ground_motion[Corr_List[i]].values,\
                    ground_motion[Corr_List[j]].values)
plt.savefig('corr_inp2.pdf', bbox_inches='tight')
```
